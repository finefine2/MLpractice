#### 데이터 전처리
데이터 전처리는 원시 데이터를 분석이나 머신러닝 모델에 적합한 형태로 변환하는 과정을 의미합니다. 전처리는 다음과 같은 이유로 중요합니다.

- 품질 향상: 잘못된 데이터, 누락된 데이터 또는 잡음이 많은 데이터는 분석 결과의 정확성을 저해할 수 있습니다.
- 효율적인 모델링: 더 깔끔하고 구조화된 데이터는 머신러닝 모델 학습에 더 유용하며 더 나은 결과를 도출할 수 있습니다.
- 분석의 용이성: 깔끔한 데이터는 데이터 분석가나 연구자가 데이터를 이해하고 분석하는 데 더 용이합니다.

##### 데이터 전처리의 주요 단계와 예시는 다음과 같습니다.

- 데이터 클렌징
    - 누락된 값 처리: 예를 들어 평균, 중앙값, 최빈값 등으로 누락된 값을 채움.
    - 이상치 탐지 및 처리: z-점수, IQR 기반 등의 방법으로 이상치를 식별하고 제거하거나 대체.
- 데이터 변환
    - 정규화: 데이터의 스케일을 [0, 1] 범위로 조정.
    - 표준화: 데이터의 평균을 0, 표준 편차를 1로 조정.
    - 원-핫 인코딩: 범주형 변수를 0과 1로 구성된 이진 벡터로 변환.
- 데이터 축소
    - 차원 축소: PCA(주성분 분석) 등의 기법을 사용하여 변수의 수를 줄임.
    - 특징 선택: 중요한 특징만 선택하여 데이터의 차원을 줄임.
- 데이터 구조 변경
    - 재구성: 피벗 테이블이나 데이터 그룹화를 사용하여 데이터 형태 변환.
    - 데이터 결합: 여러 데이터 소스에서 정보를 병합.

#### 결측값 처리 
1. 제거: 결측값이 포함된 행 또는 열을 완전히 제거합니다. 이 방법은 데이터 손실이 발생할 수 있습니다.
2. 평균, 중앙값, 최빈값으로 대체: 수치형 데이터의 경우 평균 또는 중앙값으로, 범주형 데이터의 경우 최빈값으로 결측값을 대체합니다.
3. 보간: 시계열 데이터에서 주로 사용되며, 앞뒤의 데이터를 기반으로 결측값을 추정합니다.
4. 모델을 사용한 대체: 다른 변수들을 사용하여 결측값을 가진 변수를 예측하는 모델(예: 회귀, KNN)을 학습시킨 후, 이를 사용하여 결측값을 대체합니다.

```python
# example 
data = {
    'Age': [25,30,35,None,40],
    'Salary': [50000,55000,None,62000,None]
}
df = pd.DataFrame(data) 

# 1. 결측값 제거 
df_dropna = df.dropna() 

# 2. 평균값으로 대체 
df_filled_mean = df.fillna(df.mean()) 

# 3. 중앙값으로 대체 
df_filled_median = df.fillna(df.median()) 

# 4. 최빈값으로 대체 (주로 범주형 변수에 사용) 
df_filled_mode = df['Age'].fillna(df['Age'].mode()[0])

print("데이터")
print(df)
print("\n1. 결측값 제거")
print(df_dropna)
print("\n2. 평균값으로 대체")
print(df_filled_mean)
print("\n3. 중앙값으로 대체")
print(df_filled_median)
print("\n4. 최빈값으로 대체 (주로 범주형 변수에 사용)")
print(df_filled_mode)
```

### 범주형 변수 인코딩
범주형 변수 인코딩은 머신러닝 모델을 학습하기 전에 수행해야 하는 중요한 전처리 단계 중 하나입니다. 대부분의 머신러닝 알고리즘은 숫자형 데이터만 처리할 수 있기 때문에 범주형 데이터를 적절한 숫자형 데이터로 변환해야 합니다.

#### 범주형 변수 인코딩 방법
1. Label Encoding: 범주 각각에 고유한 정수를 할당합니다. 주로 순서가 있는 범주형 데이터에 사용됩니다.
2. One-Hot Encoding: 범주 각각을 이진 특징으로 변환합니다. 이 방법은 순서가 없는 범주형 데이터에 주로 사용됩니다.


```python
from sklearn.preprocessing import LabelEncoder 
# 예제 데이터 생성
data = {
    'Age': [25, 30, 35, 40, 45],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male'],
    'Grade': ['A', 'B', 'A', 'C', 'B']
}
df = pd.DataFrame(data)

# 볌주형 변수만 선택 
categorical_cols = df.select_dtypes(include=['object']).columns 

# Label Encoding 
label_encoders = {} 
for col in categorical_cols: 
    le = LabelEncoder() 
    df[col+'_label'] = le.fit_transform(df[col])
    label_encoders[col] = le 
# one-hot encoding 
df = pd.get_dummies(df,columns=categorical_cols,drop_first=True) 
print(df)
    
```

```python
# object 타입(범주형)인 변수의 리스트 추출 
categorical_cols = train.select_dtypes(include=['object']).columns.tolist() 
categorical_cols
```

```python
# 'categorical_cols'에 해당하는 column들을 기반으로 더미변수를 생성합니다.
dum = pd.get_dummies(train,columns=categorical_cols) 
# 기존 데이터와 더미변수를 하나로 합칩니다.
train = pd.merge(train,dum,how="left") 
train.head()
```

### 이상치 처리
이상치는 다른 관측값들과 크게 다른 값을 가지는 데이터 포인트를 의미합니다. 이상치는 여러 원인(측정 오류, 입력 오류 등)으로 인해 발생할 수 있으며, 이러한 이상치가 포함된 데이터를 사용하여 모델을 학습하면 모델의 성능에 부정적인 영향을 미칠 수 있습니다. 따라서 데이터 전처리 과정에서 이상치를 식별하고 적절하게 처리하는 것이 중요합니다.

#### 이상치 탐지 방법
1. 통계적 방법: 데이터의 평균과 표준편차를 사용하여 이상치를 식별합니다. 예를 들어, 평균에서 3배의 표준편차보다 크거나 작은 값은 이상치로 간주할 수 있습니다.
2. IQR (사분위수 범위): 1사분위(Q1)와 3사분위(Q3)를 계산하고, IQR = Q3 - Q1을 구합니다. Q1 - 1.5IQR 보다 작거나 Q3 + 1.5IQR 보다 큰 값은 이상치로 간주합니다.
3. 시각적 방법: 박스 플롯, 히스토그램, 산점도 등을 사용하여 데이터를 시각화하고 이상치를 직접 확인합니다.

#### 이상치 처리 방법
- 제거: 이상치를 포함하는 행을 데이터셋에서 제거합니다.
- 대체: 이상치를 다른 값(예: 중앙값)으로 대체합니다.
- 변환: 로그 변환과 같은 기법을 사용하여 이상치의 영향을 줄입니다.

```python
# example 코드 
# 예제 데이터 생성
data = {
    'Value': [15, 14, 16, 15, 100, 14, 13, 15, 200, 16]
}
df = pd.DataFrame(data)

# IQR을 사용하여 탐지 
Q1 = df['Value'].quantile(0.25) 
Q3 = df['Value'].quantile(0.75) 
IQR = Q3 - Q1 
filter = (df['Value']>=Q1-1.5*IQR) & (df['Value']<=Q3+1.5*IQR) 
df_filtered = df[filter] 

print("Original Data") 
print(df) 
print("\n Data after removing outliers")
print(df_filtered)
```

위 방법들 중 1~3의 방법을 적용하여 모델의 정확도를 향상해보고자 합니다.

#### 파생변수
파생변수란 기존에 가지고 있는 변수들을 활용하여 새로운 변수를 생성하는 것을 의미합니다. 파생변수 생성은 데이터의 정보를 더욱 풍부하게 만들어 머신러닝 모델의 성능을 향상시키는 데 도움을 줄 수 있습니다.

##### 파생변수 생성
- 파생변수 생성의 이유:
1. **정보 가치 향상**: 원본 데이터셋에 없는 새로운 정보나 패턴을 포착하기 위해 파생변수를 만듭니다.
2. **데이터의 해석력 향상**: 비즈니스 문제나 도메인 지식을 기반으로 파생변수를 만들어 데이터의 해석력을 높일 수 있습니다.
3. **모델 성능 향상**: 더 많은 정보를 제공하여 모델의 예측 성능을 향상시킬 수 있습니다.
4. **데이터의 차원 축소**: 여러 변수의 정보를 합쳐 하나의 파생변수로 축소할 수 있습니다.

```python
import numpy as np 
# 파생변수 생성 및 주석
# 1. 파생변수 생성을 위해 음수값 양수로 전환
# 일부 변수들에서의 일수는 음수로 표시되어 있어 계산의 편의성을 위해 양수로 전환합니다.
negative_list = ['DAYS_EMPLOYED', 'DAYS_BIRTH', 'begin_month']
for i in negative_list:
    train[f"{i}"] = -train[f"{i}"]

# 2. 사용자의 나이 계산
# 사용자가 태어난 지 몇 일이 지났는지를 기반으로 나이를 계산합니다.
train['Age'] = abs((train['DAYS_BIRTH']) / 365)

# 3. 사용자의 고용된 연수 계산
# 얼마나 오래 전에 고용되었는지를 나타내는 변수입니다.
train['Years_Employed'] = train['DAYS_EMPLOYED'].apply(lambda x: -x / 365 if x < 0 else 0)  # 양수값은 무직자로 간주하고 0 처리
train.loc[train['DAYS_EMPLOYED'] < 0, 'DAYS_EMPLOYED'] = 0

# 4. 사용자의 연락 수단 수 계산
# 사용자가 가지고 있는 연락 수단의 총 수를 계산합니다.
train['Contact_Count'] = train['FLAG_MOBIL'] + train['work_phone'] + train['phone'] + train['email']

# 5. 소득 대비 자녀 수
# 각 사용자의 소득에 대한 자녀 수의 비율을 계산합니다.
train['income_per_child'] = train['income_total'] / (train['child_num'] + 1)

# 6. 부양가족 수당 소득
# 사용자의 총 소득을 부양가족 수로 나누어 한 부양가족 당 평균 소득을 계산합니다.
train['income_per_family'] = train['income_total'] / (train['family_size'])

# 7. 고용전 일수
# 사용자가 처음으로 고용되기 전까지의 전체 일수를 계산합니다.
train['before_EMPLOYED'] = train['DAYS_BIRTH'] - train['DAYS_EMPLOYED']

# 8. 고용되기 전까지의 시간 대비 총 소득
# 사용자의 총 소득을 고용되기 전까지의 일수로 나눈 값을 계산합니다.
train['income_total_befofeEMP_ratio'] = train['income_total'] / train['before_EMPLOYED']

# 9. 사용자가 처음으로 고용되기 전까지의 월 수
# 사용자가 처음으로 고용되기 전까지의 월 수를 계산합니다.
train['before_EMPLOYED_m'] = np.floor(train['before_EMPLOYED'] / 30) - ((np.floor(train['before_EMPLOYED'] / 30) / 12).astype(int) * 12)

# 10. 가장의 명수
# 부양 가족 수에서 자녀의 수를 뺀 값은 경제 활동을 할 수 있는 가장의 수를 나타냅니다.
train["active_parents"] = train["family_size"]-train["child_num"]
train.loc[train['active_parents'] < 0, 'active_parents'] = 0

# 출력
train.describe()
```

#### Kmeans 클러스터링 학습
KMeans 클러스터링은 비지도 학습 알고리즘 중 하나로, 주어진 데이터를 'K'개의 클러스터로 그룹화하는 알고리즘입니다. 여기서 'K'는 사용자가 지정해야 하는 클러스터 수를 나타냅니다.

##### KMeans의 작동 원리
1. 초기 'K'개의 클러스터 중심을 데이터 포인트에서 무작위로 선택하거나 다른 방법으로 초기화합니다.
2. 각 데이터 포인트를 가장 가까운 클러스터 중심에 할당합니다.
3. 각 클러스터의 중심을 해당 클러스터의 데이터 포인트 평균으로 업데이트합니다.
4. 클러스터 중심의 이동이 없을 때까지 2번과 3번 단계를 반복합니다.

##### KMeans의 주요 특징
- 초기 중심값의 선택: KMeans의 결과는 초기 중심값의 선택에 따라 달라질 수 있으므로, 여러 번 실행하거나 초기화 방법을 변경해보는 것이 좋습니다.
- 거리 측정: 통상적으로 유클리디안 거리를 사용하여 데이터 포인트와 클러스터 중심 간의 거리를 측정합니다.
- 클러스터 수 'K'의 선택: 'K'값은 사용자가 지정해야 하는데, 이를 선택하기 위한 방법으로 Elbow 방법 등이 있습니다.

```python
from sklearn.cluster import KMeans 
import os 
import matplotlib.pyplot as plt 
os.environ['OMP_NUM_THREADS'] = '1' # warning 메세지 비활성화

# 예제 데이터 생성
data = {
    'x': [25, 34, 22, 27, 33, 33, 31, 22, 35, 34, 67, 54, 57, 43, 50, 57, 59, 52, 65, 47, 49],
    'y': [79, 51, 53, 78, 59, 74, 73, 57, 69, 75, 51, 32, 40, 47, 53, 36, 35, 58, 59, 50, 25]
}

df = pd.DataFrame(data) 

plt.scatter(df['x'],df['y'])
plt.xlabel('X Value') 
plt.ylabel('Y Value')
plt.title('Visualization of raw data')
plt.show() 

# KMeans 클러스터링 
kmeans = KMeans(n_clusters=3,n_init=10) 
df['cluster'] = kmeans.fit_predict(df) 

# 클러스터링 결과 시각화 
plt.scatter(df['x'],df['y'],c=df['cluster'],cmap='rainbow') 
plt.xlabel('X Value') 
plt.ylabel('Y Value') 
plt.title('KMeans Clustering') 
plt.show()
```

**범주형 데이터를 수치형 데이터로 변환**

KMeans 알고리즘은 수치형 데이터에 기반한 거리 기반의 클러스터링 알고리즘이므로, 범주형 데이터를 적절한 수치형 데이터로 변환해야 합니다. 여기서는 각 범주를 숫자로 매핑하는 간단한 방법을 사용하겠습니다.

```python
# 범주형 변수 선택 
kmeans_cols = ['income_type', 'edu_type']

# label encoding 
label_encoders = {} 
for col in kmeans_cols: 
    le = LabelEncoder() 
    kmeans_data[col] = le.fit_transform(kmeans_data[col]) 
    label_encoders[col] = le 
kmeans_data[['income_type','edu_type']].head()
```

**KMeans 알고리즘 적용 및 WCSS 값 계산**

이제 변환된 수치형 데이터를 사용하여 KMeans 알고리즘을 적용하고, 다양한 k 값에 대한 WCSS 값을 계산하겠습니다. WCSS는 클러스터 내의 제곱 거리의 합을 나타내며, 이 값을 사용하여 최적의 k 값을 선택하는 데 도움을 받을 수 있습니다.

WCSS 값이 급격히 감소하는 지점, 즉 "elbow point"를 찾아서 최적의 클러스터 개수를 결정할 수 있습니다. 여기서는 k의 범위를 1부터 10까지로 설정하고 WCSS 값을 계산해보겠습니다.

```python
# 변환된 수치형 데이터 추출
kmean_target = kmeans_data[['income_type','edu_type']]
# 다양한 k 값에 대한 WCSS 값을 저장할 리스트 초기화
wcss = [] 
# k = 1부터 10까지 KMeans 알고리즘 적용 및 WCSS 값 계산
for i in range(1,11): 
    kmeans = KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0) 
    kmeans.fit(kmean_target) 
    wcss.append(kmeans.inertia_) 
wcss
```

**WCSS 값에 대한 그래프 시각화**

이제 WCSS 값을 그래프로 시각화하여 최적의 클러스터 개수 k 값을 확인하겠습니다. 그래프에서 "elbow point"를 찾아서 최적의 k 값을 결정할 수 있습니다.

```python
import matplotlib.pyplot as plt 
# WCSS visualization 
plt.figure(figsize=(10,5)) 
plt.plot(range(1,11),wcss,marker='o',linestyle='--')
plt.title('Elvow Method')
plt.xlabel('Number of clusters') 
plt.ylabel('WCSS') 
plt.grid(True) 
plt.show()
```

이제 WCSS 값을 기반으로 한 Elbow Method 그래프를 확인할 수 있습니다. 그래프에서 "elbow point"는 k=4 근처에 위치하는 것처럼 보입니다. 따라서 4를 최적의 클러스터 개수로 선택할 수 있습니다.
이 그래프를 기반으로 하는 것 외에도 실제 업무 환경에서는 도메인 지식, 다른 평가 지표 및 추가적인 실험을 통해 최적의 k 값을 결정합니다.

- GridSearchCV는 하이퍼파라미터 튜닝 작업을 수행하기 위한 함수입니다.
- xgb_grid는 최적의 파라미터를 찾기 위해 테스트할 모델입니다.

#### Feature Importances 및 변수 Customizing
**모델의 성능은 불필요한 변수를 제거함으로써 향상될 수 있으며 여러 가지 이유가 있습니다:**

1. 차원의 저주 (Curse of Dimensionality):
고차원의 데이터 공간에서는 데이터 포인트 간의 거리가 크게 증가하기 때문에, 예측 모델이 올바른 판단을 내리기가 어려워집니다. 이는 특히 군집화, 최근접 이웃 검색과 같은 거리 기반 알고리즘에서 중요합니다. 불필요한 변수를 제거함으로써 차원을 줄이면, 데이터 포인트 간의 관계를 더 쉽게 파악할 수 있게 됩니다.

2. 과적합 (Overfitting):
불필요한 변수가 많을수록 모델은 훈련 데이터에 대한 불필요한 패턴까지 학습할 위험이 있습니다. 이로 인해 테스트 데이터나 실제 환경에서의 성능이 저하될 수 있습니다. 변수를 제거하면 모델의 복잡성이 감소하고, 과적합을 방지할 수 있습니다.

3. 계산 효율성:
변수의 수가 줄어들면 학습, 예측, 검증 과정의 계산 비용이 줄어듭니다. 이로 인해 모델의 학습과 추론 속도가 빨라집니다.

4. 모델의 해석성 (Interpretability):
중요하지 않은 변수가 제거되면, 남아있는 변수들 사이의 관계가 더 명확해집니다. 따라서 모델의 예측을 해석하기가 더 쉬워집니다.

5. 다중공선성 (Multicollinearity):
변수들 사이에 강한 상관관계가 있을 경우, 다중공선성이 발생할 수 있습니다. 이는 회귀 모델에서 계수의 추정치가 불안정해질 수 있게 만듭니다. 불필요한 변수를 제거함으로써 다중공선성 문제를 줄일 수 있습니다.

6. 노이즈 감소:
불필요한 변수는 종종 데이터의 노이즈를 증가시키는 원인이 됩니다. 이러한 변수를 제거함으로써 노이즈가 감소하고, 모델의 일반화 성능이 향상될 수 있습니다.

7. 특성 선택의 중요성:
특성 선택을 통해 가장 중요한 변수들만을 사용하면, 해당 변수들이 예측에 얼마나 큰 영향을 미치는지 더 잘 이해할 수 있습니다. 요약하면, 불필요한 변수를 제거함으로써 모델의 성능은 물론, 학습 속도, 모델의 해석성 등 여러 면에서 이점을 얻을 수 있습니다.

이제 모델의 Feature Importances를 추출하고 그에 맞게 변수를 조절하며 모덜 성능을 개선해보겠습니다.
